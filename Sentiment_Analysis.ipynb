# Sentiment Analysis Project

### Goal
The goal of this project is to analyze a sentiment dataset containing user posts across multiple platforms.  
We aim to clean the data, explore trends, and gain insights into user behavior, sentiment distribution, and engagement patterns (likes, retweets, activity).  

### Tools Used
- Python (Pandas, NumPy)
- Visualization: Matplotlib, Seaborn
- Jupyter Notebook


```python
import pandas as pd

print("Setup ready ")
```

```python
file_path = r"C:\Users\sivanisree\Downloads\Data Set For Task.csv\Data_Set_For_Task\3) Sentiment dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
duplicates_total = df.duplicated().sum()
print("Total duplicate rows:", duplicates_total)

# Check duplicates only in the 'Text' column (since the same post might repeat)
duplicates_text = df["Text"].duplicated().sum()
print("Duplicate posts (Text column):", duplicates_text)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# 2. Standardize Sentiment values (lowercase + remove extra spaces)
df['Sentiment'] = df['Sentiment'].str.lower().str.strip()

# 3. Standardize Platform values (capitalize first letter, e.g., "twitter", "instagram")
df['Platform'] = df['Platform'].str.title().str.strip()

# 4. Optional: remove unwanted index columns if theyâ€™re not needed
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], errors='ignore')

# Check the first 5 rows again
print(df.head())
```

```python
# Extract useful time features
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Month_Name'] = df['Timestamp'].dt.month_name()
df['Day'] = df['Timestamp'].dt.day
df['Weekday'] = df['Timestamp'].dt.day_name()
df['Hour'] = df['Timestamp'].dt.hour

# Quick check
print(df[['Timestamp','Year','Month','Month_Name','Day','Weekday','Hour']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top_sentiments = df['Sentiment'].value_counts().head(10).index

plt.figure(figsize=(8,5))
sns.countplot(x='Sentiment', data=df[df['Sentiment'].isin(top_sentiments)],
              order=top_sentiments, palette='muted')
plt.title("Top 10 Most Common Sentiments")
plt.xticks(rotation=45)
plt.show()
```

## Top 10 Most Common Sentiments  

- The most frequent sentiment is **gratitude**, followed by **hope** and **joy**.  
- Overall, the dataset shows a stronger presence of **positive emotions** compared to negative ones.  


```python
sentiment_map = {
    "joy": "Positive", "elation": "Positive", "happiness": "Positive", "contentment": "Positive",
    "serenity": "Positive", "fulfillment": "Positive", "gratitude": "Positive", "hope": "Positive",
    "success": "Positive", "love": "Positive", "peace": "Positive", "inspiration": "Positive",
    "compassion": "Positive", "optimism": "Positive", "confidence": "Positive",

    "grief": "Negative", "sadness": "Negative", "despair": "Negative", "anger": "Negative",
    "fear": "Negative", "disgust": "Negative", "loneliness": "Negative", "betrayal": "Negative",
    "melancholy": "Negative", "regret": "Negative", "bitterness": "Negative", "loss": "Negative",
    "frustration": "Negative", "hate": "Negative",

    "neutral": "Neutral", "indifference": "Neutral", "contemplation": "Neutral",
    "nostalgia": "Neutral", "curiosity": "Neutral", "surprise": "Neutral"
}

# Step 2: Apply mapping
df['Sentiment_Group'] = df['Sentiment'].str.lower().map(sentiment_map)

# Step 3: Fill unmapped ones as Neutral
df['Sentiment_Group'] = df['Sentiment_Group'].fillna("Neutral")

# Step 4: Check distribution
print(df['Sentiment_Group'].value_counts())
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Hour', data=df, palette='coolwarm')
plt.title("Posts per Hour of Day")
plt.show()

# 4. Sentiment by platform
plt.figure(figsize=(6,4))
sns.countplot(x='Platform', hue='Sentiment_Group', data=df, palette='Set2')
plt.title("Sentiment by Platform")
plt.xticks(rotation=45)
plt.show()
```

## Posts per Hour of Day

- Posting activity is highest between **6â€“10 PM**, showing that users engage more during evenings.

- Very few posts are made after midnight, suggesting lower late-night activity.

## Sentiment by Platform

- **Instagram** posts are dominated by positive emotions.

- **Twitter** shows a more balanced mix of positive, neutral, and negative sentiments.

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Create a Weekday column
df['Weekday'] = df['Timestamp'].dt.day_name()

# Now plot again
plt.figure(figsize=(8,4))
sns.countplot(x='Weekday', data=df,
              order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
              palette='muted')
plt.title("Posts per Weekday")
plt.xticks(rotation=45)
plt.show()
```

## Posts per Weekday

- Posting activity is **highest** on Friday and Saturday, showing stronger engagement before weekends.

- Midweek (Tuesdayâ€“Thursday) shows moderate posting levels.

- On Wednesday is the **lowest** posting

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Platform', data=df, palette='muted')
plt.title("Posts per Platform")
plt.show()
```

## Posts per Platform

- **Instagram** is the most active platform, contributing nearly 40% of posts.

- **Twitter** and **Facebook** follow, but with smaller shares.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Year", hue="Sentiment_Group", palette="coolwarm")
plt.title("Yearly Sentiment Trend")
plt.xlabel("Year")
plt.ylabel("Number of Posts")
plt.show()
```

## Yearly Sentiment Trend

- From **2010** to **2018**, posts were relatively low.

- After **2020**, post activity increased sharply, possibly due to wider social media usage and global events.

- Positive sentiments consistently outnumber negative ones across the years. However, in **2019**, negative sentiments got high.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Sentiment_Group', y='Likes', data=df, palette='pastel')
plt.title("Likes Distribution by Sentiment")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x='Sentiment_Group', y='Retweets', data=df, palette='muted')
plt.title("Retweets Distribution by Sentiment")
plt.show()
```

## Likes Distribution by Sentiment

- Positive posts receive the highest number of likes on average.

- Neutral posts get moderate engagement, while negative posts receive fewer likes.

## Retweets Distribution by Sentiment

- **Negative posts** show slightly higher retweet counts, suggesting that controversial/emotional content spreads faster.

- **Positive posts** still receive steady engagement but are less viral compared to negative ones.

```python
top_likes = df.sort_values(by='Likes', ascending=False).head(5)[['Text','Likes','Sentiment_Group']]
top_retweets = df.sort_values(by='Retweets', ascending=False).head(5)[['Text','Retweets','Sentiment_Group']]

print("Top 5 Most Liked Posts:\n", top_likes)
print("\nTop 5 Most Retweeted Posts:\n", top_retweets)
```

## Insights from EDA
- **Sentiment Distribution**: Most of the posts are *Positive*, followed by *Neutral*. Negative posts are the least common.  
- **Posts per Platform**: Instagram has the highest activity compared to other platforms.  
- **Posts per Weekday**: Engagement is higher during weekdays (especially Monday and Tuesday).  
- **Retweets by Sentiment**: Positive posts tend to receive slightly more retweets compared to negative or neutral ones.  
- **Likes by Sentiment**: Positive posts clearly get more likes on average.
